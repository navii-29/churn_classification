```python
import pickle
import pandas as pd
import numpy
from tensorflow.keras.models import load_model
```

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

from tensorflow.keras.models import load_model
model = load_model("tuned_model.keras")
```

```python
with open('gender_label-2.pkl','rb') as file:
    gender_label = pickle.load(file)


with open('scale.pkl','rb') as file:
    scaler = pickle.load(file)

with open('ohe_geo.pkl','rb') as file:
    ohe_geo = pickle.load(file)


```

```python
input_data = {

    'CreditScore':600, 
    'Geography': 'France',
    'Gender': 'Male', 
    'Age': 40, 
    'Tenure':3, 
    'Balance': 60000, 
    'NumOfProducts': 2, 
    'HasCrCard': 1,
    'IsActiveMember': 1, 
    'EstimatedSalary': 50000, 


}
```

```python
input_df = pd.DataFrame([input_data])
```

```python
geo_encoded = ohe_geo.transform([[input_data['Geography']]]).toarray()
```

```python
geo_encoded
```

```python
geo_data = pd.DataFrame(geo_encoded,columns = ohe_geo.get_feature_names_out(['Geography']))
```

```python
geo_data
```

```python
ohe_df = pd.concat([input_df,geo_data],axis = 1)
```

```python
ohe_df.shape
```

```python
ohe_df['Gender'] = gender_label.transform(ohe_df['Gender'])
```

```python
ohe_df.drop('Geography', axis = 1, inplace = True)
```

```python
scaled_df = scaler.transform(ohe_df)
```

```python
prediction  = model.predict(scaled_df)

if prediction[0][0] > 0.5:
    print('customer is likely to churn')
else:
    print('customer is likely to stay')
```
